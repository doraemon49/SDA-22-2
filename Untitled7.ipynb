```python
import pandas as pd
import seaborn as sns
import numpy as np
from pandas import DataFrame
from pandas import Series
from scipy import stats
from scipy.stats import kstest
from scipy.stats import bartlett
from scipy.stats import levene

data = pd.read_csv("Life Expectancy Data.csv")
data.head()
```

```python
#null 확인 및 제거
np.sum(pd.isnull(data))
d1 = data["Country"]
d2 = data["Life_expect"]
data2 = pd.concat([d1,d2],axis=1)
data2 = data2.dropna()
print(np.sum(pd.isnull(data2)))
print(data2)
```

```python
#country 별 life_expect
grouped_cToe = data2["Life_expect"].groupby(data2["Country"])
print("size: \n",grouped_cToe.size())
grouped_cToe_mean = grouped_cToe.mean()
print("mean: \n",grouped_cToe_mean)
```

```python
from scipy import stats
print(stats.ttest_ind(grouped_cToe_mean[1]))
```
